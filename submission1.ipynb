- Using pandas data can be cleaned, transformed, manipulated, & analyzed.
- The basic data structures of pandas are Series and DataFrame.

1. Series object: Series is a one dimensional labelled array.
2. DataFrame: A DataFrame in pandas is two-dimensional tabular data structure.
- It consists of rows and columns represents different features.
- rows represents records or entries.
- index represents label of rows.
