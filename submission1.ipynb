Exercise 1 : Describe Any 10 AI Libraries
Description: This tutorial provides an overview of 10 popular AI libraries, focusing on their functionalities, applications, and key features.

1. TensorFlow

- TensorFlow is an open-source AI library developed by Google for building and deploying machine learning and deep learning models.

- Functionality: Neural networks, deep learning, numerical computation

- Applications: Image recognition, speech recognition, NLP, recommender systems

- Key Features:

    Supports CPUs, GPUs, and TPUs

    Scalable for large datasets

    TensorFlow Lite for mobile and embedded devices

2. PyTorch

- PyTorch is an open-source deep learning framework developed by Meta (Facebook), popular in research and academia.

- Functionality: Deep learning, neural networks

- Applications: Computer vision, NLP, reinforcement learning

- Key Features:

    Dynamic computation graphs

    Easy debugging and flexibility

    Strong community support

3. Keras

- Keras is a high-level neural network API that runs on top of TensorFlow.

- Functionality: Simplified deep learning model creation

- Applications: Image classification, text classification, prototyping AI models

- Key Features:

    User-friendly and intuitive

    Fast model building and experimentation

    Supports CNNs and RNNs

4. OpenCV

- OpenCV (Open Source Computer Vision) is a library focused on real-time computer vision.

- Functionality: Image and video processing

- Applications: Face detection, object tracking, video analytics

- Key Features:

    Supports multiple programming languages

    Optimized for real-time applications

    Large collection of vision algorithms

5. Scikit-Learn

- Scikit-Learn is a Python library for traditional machine learning.

- Functionality: Data mining and data analysis

- Applications: Classification, regression, clustering, prediction systems

- Key Features:

    Easy-to-use API

    Built-in datasets and evaluation tools

    Integrates well with NumPy and Pandas

6. NLTK (Natural Language Toolkit)

- NLTK is one of the oldest and most widely used NLP libraries in Python.

- Functionality: Text processing and linguistic analysis

- Applications: Chatbots, sentiment analysis, text classification

- Key Features:

    Tokenization, stemming, parsing

    Large set of linguistic datasets

    Ideal for learning and research

7. SpaCy

- SpaCy is a modern NLP library designed for production use.

- Functionality: Advanced natural language processing

- Applications: Information extraction, named entity recognition, text summarization

- Key Features:

    Fast and efficient

    Pre-trained language models

    Industrial-strength performance

8. GPT-3 (OpenAI)

- GPT-3 is a powerful language model developed by OpenAI and accessed via API.

- Functionality: Text generation and language understanding

- Applications: Chatbots, content writing, code generation

- Key Features:

    Few-shot learning

    Human-like text generation

    Supports multiple languages

9. IBM Watson

- IBM Watson is an enterprise AI platform offering multiple AI services.

- Functionality: AI-powered analytics and automation

- Applications: Healthcare, customer support, business intelligence

- Key Features:

    Natural language understanding

    Speech-to-text and text-to-speech

    Cloud-based AI solutions

10. RapidMiner

- RapidMiner is a data science platform that supports AI and machine learning workflows.

- Functionality: Data preparation, model building, evaluation

- Applications: Business analytics, predictive modeling

- Key Features:

    Drag-and-drop interface

    No-code/low-code platform

    Supports machine learning and deep learning

HomeWork : implement Boxploat, Scatter Plot and Line Plot with suitable data

```python
# Boxplot

plt.figure(figsize=(6,4))
df.boxplot(column='price', by='class')
plt.xlabel('Class')
plt.ylabel('Price')
plt.title('Box Plot: Price by Class')
plt.suptitle('')
plt.show()
```

```python
# Scatter Plot

plt.figure(figsize=(6,4))
plt.scatter(df['duration'], df['price'], alpha=0.5)
plt.xlabel('Duration (hours)')
plt.ylabel('Price')
plt.title('Scatter Plot: Duration vs Price')
plt.show()
```

```python
# Line Plot

avg_price = df.groupby('days_left')['price'].mean()

plt.figure(figsize=(6,4))
plt.plot(avg_price.index, avg_price.values)
plt.xlabel('Days Left')
plt.ylabel('Average Price')
plt.title('Line Plot: Average Price vs Days Left')
plt.show()
```
