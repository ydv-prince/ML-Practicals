```python
def search(l1, target):
    indexes = []
    for i in range(len(l1)):
        if target == l1[i]:
            indexes.append(i)
    if indexes:
        return indexes
    else:
        return -1

l1 = [2,4,5,8,8,9]
search(l1,0)
```
